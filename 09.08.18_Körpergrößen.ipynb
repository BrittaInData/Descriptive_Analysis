# Verteilung von Körpergrößen

```python
# Module importieren
import numpy as np
import matplotlib.pyplot as plt
import scipy, scipy.stats

# Diagramm im Stil 'seaborn' plotten
plt.style.use('seaborn')
```

```python
# Für die Generierung der Daten für das Histogramm erzeugen wir Zufallswerte mit Numpy. 
# Um bei jedem Programmdurchlauf die selben Werte zu erhalten, 
# versehen wir den Zufallsgenerator mit dem Startwert 1.
np.random.seed(1)

# Erwartungswert bzw. Mittelwert mu und Standardabweichung sigma definieren
mu = 170
sigma = 8
 
# Normalverteilte Zufallswerte generieren, wird durch die Funktion ermöglicht
x = sigma * np.random.randn(10000) + mu

# Histogramm zeichnen, normalisieren mit dem Parameter density und die Anzahl der Balken auf 100 begrenzen
plt.figure(figsize=(8, 6))
plt.hist(x, 100, density=True)
plt.show()
```

```python
# Beschriftungen setzen
plt.xlabel('Körpergröße')
plt.ylabel('Wahrscheinlichkeit')
plt.title('Normalverteilung von Körpergrößen')

# Werte für mu und sigma links oben in  Grafik schreiben:  
# erst die Koordinaten,, dann in LaTeX Schreibweise "mu" + "sigma".
# LaTeX ist ein Textsatzsystem.
plt.text(190, 0.05, '$\mu=170,\ \sigma=8$')

# Bild anzeigen
plt.hist(x, 100, density=True)
plt.show()
```
