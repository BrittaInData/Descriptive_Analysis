```python
# Module importieren
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import scipy.stats as stats

# Sp√§tere Diagramme im Stil 'seaborn' plotten
plt.style.use('seaborn')

# Erwartungswert bzw. Mittelwert mu und Standardabweichung sigma definieren
mu = 170
sigma = 8

# Normalverteilte Zufallswerte mit Sigma und mu
x = sigma * np.random.randn(100000) + mu 

# randn: Standardnormalverteile Zufallsvariablen
# Echte Normalverteilung mit mu und sigma
z = scipy.linspace(120,220,300)
pdf = scipy.stats.norm.pdf(z,mu,sigma)

# Histogramm zeichnen, normalisieren mit dem Parameter density und die Anzahl der Balken auf 100 begrenzenp
plt.hist(x, 22, density=True, facecolor='blue')
plt.plot(z, pdf)
plt.show();
```
